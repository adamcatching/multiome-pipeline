# GOAL(S)

- Get all the results of the DGE together to look at on Web Gestalt

# IMPORTS

```python
import pandas as pd
import os
import pprint
```

# LOAD DATA

```python
directory_path = 'output/DGE_results_df-pass_thresholds'
```

# LOOP THRU FOLDER AND CONCAT

```python
# List all files in the directory
files = os.listdir(directory_path)

# print('files =', files)
# print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

for file in files:
    print(file)
```

```python
# Filter out only CSV files
csv_files = [file for file in files if file.endswith('.csv')]
csv_files
```

```python
# Initialize an empty list to hold DFs
df_list = []
```

```python
# Loops thru CSV files to read into DFs
for csv_file in csv_files:
    print('csv_file =', csv_file)
    file_path = os.path.join(directory_path, csv_file)
    df = pd.read_csv(file_path)
    df_list.append(df)
    
df_list
```

```python
# Concat all DFs
big_df = pd.concat(df_list, ignore_index=True)
big_df
```
